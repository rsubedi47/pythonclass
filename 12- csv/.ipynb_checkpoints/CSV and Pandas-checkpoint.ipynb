```python
## defining_index_col()

import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  # sn is origin, ie. x-axis, sn is index column
df  
```

```python
# loc
#iloc = index localizer

## iloc(rows)

import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  # sn is origin, ie. x-axis, sn is index column
df.iloc[1:5]
```

```python
## iloc (rows and columns)
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  # sn is origin, ie. x-axis, sn is index column
df.iloc[1:5,0:2]  ## 0 dekhi column 2 index samma
```

```python
#### loc######## = index thaha navayepani vayo

# on the basis of index column_i.e yaha = sn ko basis ma 

import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  
df.loc[2:5]    # serian number sn 2 dekhi 5 samma -- as hamile sn lai index_col banako thiyau
```

```python
## loc on the basis of name as index_col
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'name')  
df.loc['Shyam':'Sabina']        
```

```python
# usecols

import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn',usecols= ['sn','name','address'])  
df
```

```python

## name lai list ma nikaleko
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  
list(df['name'])
```

```python
# filtering

# name
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  
df[df['name'] == 'Ram']  # remember the brackets
```

```python

# filtering with age

import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  
df[df['age'] > 30]   # filter
```

```python
####### multiple conditions ##

import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  
df[(df['age']>30)  &  (df['address']== 'Kathmandu')]   # remember the brackets
```

```python
####### multiple conditions ##   and saving  to a new file
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')  
data = df[(df['age']>30)  &  (df['address']== 'Kathmandu')]   # remember the brackets
data.to_csv('new_data.csv')   # new file saved
```
